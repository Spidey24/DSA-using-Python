```python
#list 
# Similar to Vector in C++ or Arraylist in Java
l = [1,2,3,'ab',2.3]
print(type(l),'\n',l)

# Traversing a List 
print('\n============Traversal===========\n')
l = ['A','l','f','r','e','d']
for a in l:
  print(a,end='\t')
print('\n==========Joining list===========\n')
#Joining lists 
l1 = [1,2,3]
l2 = [4,5,6]
print(l1+l2)
print('\n==========Replicating=============\n')
#Replicating Lists
l3 = l1 * 3 
print(l3)
print('\n========Slicing list============\n')
#Slicing a list 
# Similar to String Slicing
l4 = [10,12,14,16,20,22,24,28,30,32,36,38,40] 
s = l4[0:3]
t = l4[-7:-1] #Negative slicing
u = l4[0:10:2] #Step slicing
v = l4[0:2]=['ten','twelve','fourteen'] # We can use string slicing to modify specific elements of list
print(s,end='\n')
print(t,end='\n')
print(u,end='\n')
print(v,end='\n')
print(l4)
print('===============================')
```

```python
# tuple exception is it's immutable 
t = (1,2)
print(type(t),'\n',t)
#Traversing a tuple 
print('Traversing tuple------------------')
t = ('C','o','d','i','n','g')
for a in t:
  print(a)

#Joining tuple
print('\nJoining tuple-------------------')
tp1=(1,2,3)
tp2=(4,5,6)
tp3=tp1+tp2
print(tp3)

#==============Remember====================
# tp1 + (7)
# Python will return an error : TypeErrror : can only concatenate tuple (not 'int') to tuple

#Repeating tuple
print('\nReplicating tuple---------------------')
tpl = tp1 * 3
print(tpl) 

#Slicing Tuple 
print('\nSlicing tuple--------------------------')
tp4=(10,20,22,20,33,45,56,64,76,88,91,93,94,100)
seq = tp4[3:7]
print(seq)
seq1=tp4[0:11:2]
print(seq1)

#Unpacking tuples 
# Creating a tuple from a set of values is called packing and it's reverse i.e, creating individual values from a tuple's elements is called unpacking.
print('\nUnpacking tuple---------------------')
t = (1,2,3,'a','b','c')
a,b,c,d,e,f = t
print(a,'==',b,'==',c,'==',d)
```

```python
# mapping data type (dictionary)
d = {'a':1,'b':2}
print(type(d),'\n',d)
# Travering 
print('Traversal=======================>')
dictionary1 = {1:'number','a':'string',8.1 : 'float', (2,3) : "tuple" }
for key in dictionary1:
  print(key,':',dictionary1[key])

#Adding elements 
print('\nAddition of elements===========>')
emp = {'eid' : 'A101', 'ename' : 'John Doe' , 'designation' : 'SR dev' , }
print(emp)
emp['age'] = 28
print(emp)

#Updation 
print('\nUpdation of existing elements============>')
emp['eid'] = 'B101'
print(emp)

#Deletion
print('\nDeleting elements==================>')
#Here we have 2 methods; One is del and other is pop
# del method 
print('\n-----------del method--------------')
del emp['age']
print(emp)

# If key doesn't exist which we are going to delete then Python returns error
# del emp['dept']

# pop method 
print('\n-----------pop method--------------')
dictionary1.pop('a')
print(dictionary1)

# Pop allows user to specify what to display when the given key does not exist
# dictionary1.pop(99,'No key found')

#Existence of key
print('\nExistence of key=======================>')
a = 1 in dictionary1
print(a)
b = 'salary' not in dictionary1
print(b)
```
