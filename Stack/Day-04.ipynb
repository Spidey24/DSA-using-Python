### Stack
---
* A linear data structure that allows insertion and deletion operations from one end i.e., top
* LIFO (Last In Fast Out)
* Operation performed are as follows :
    * Push (Add elements in stack) 
    * Pop (Deleting elemets in stack)
    * Top (Points towards top element in stack)
    * isEmpty (Returns true if stack is empty)

```python
# Using list 
stack = []
def push_stack():
    element = input('Enter the element:\n')
    stack.append(element)
    print(stack)
def pop_stack():
    if not stack:
        print('Stack is empty')
    else:
        p = stack.pop()
        print('Element removed:',p)
        print(stack)
def top_stack():
    if not stack:
        print('Stack is empty')
    else:
        print(stack[-1])
def isEmpty_stack():
    if len(stack) == 0:
        print('isEmpty is True')
    else:
        print('isEmpty is False')
while True:     
    print()
    choice = int(input('Select the operations to perform on stack\n 1:Push\t2:Pop\t3:Peek\t4:isEmpty\t5:Quit\n'))
    if choice == 1:
        push_stack()
    elif choice == 2:
        pop_stack()
    elif choice == 3:
        top_stack()
    elif choice == 4:
        isEmpty_stack()
    elif choice == 5:
        break
    else:
        print('Choose something between given choices only')
```
