### Queue
---
* It is a linear data structure it's open at both ends , One end is always used for insertion (enqueue) and the other is used for deletion (dequeue).
* FIFO (First-In-First-Out) 
* Operations performed :
    * enqueue (Insertion of data)
    * dequeue (Deletion of data)
    * isFull (Is queue full of memory)
    * isEmpty (Is queue empty of data)


```python
# Using list 
queue = []
n = int(input('Enter the size of queue:\n'))
def enqueue():
    if len(queue) < n:
        element = input('Enter the element:\n')
        queue.append(element)
        print(queue)
    else:
        print('Queue is Fully occupied')
def dequeue():
    if not queue:
        print('Queue is empty')
    else:
        e = queue.pop(0) # Otherwise it will work as stack
        print(e,'is removed from Queue')
def isFull():
    if len(queue) >= n:
        print('Queue is full')
    else:
        print("Queue memory isn't yet full")
def isEmpty():
    if len(queue) == 0:
        print('Queue is empty')
    else:
        print('Queue is not empty')
while True:
    choice = int(input('Select from following operations for queue\n1.enqueue\t2.dequeue\t3.isFull\t4.isEmpty\t5.Quit\n'))
    if choice == 1:
        enqueue()
    elif choice == 2:
        dequeue()
    elif choice == 3:
        isFull()
    elif choice == 4:
        isEmpty()
    elif choice == 5:
        break
    else:
        print('Enter a valid choice from above chocies only!!')
```

```python
# Using Module
from collections import deque

class Queue:
    def __init__(self,max_size):
        self.queue = deque()
        self.max_size = max_size

    def enqueue(self, item):
        if len(self.queue) < self.max_size :
            self.queue.append(item)
            print(self.queue)
        else:
            print('Queue is full\n')

    def dequeue(self):
        if len(self.queue) != 0:
            e = self.queue.popleft()
            print(e,'is removed from queue')
        else:
            print('Queue is empty')

    def isfull(self):
        if len(self.queue) >= self.max_size :
            print('Queue is full')
        else:
            print('Queue is not full')

    def is_empty(self):
        if len(self.queue) == 0:
            print('Queue is empty')
        else:
            print('Queue is not empty')

    def size(self):
        print( len(self.queue),'is current size of queue\n',self.max_size,'is maximum size of queue')

n = int(input('Enter the size of queue\n'))
my_queue = Queue(n)
print('\nEnqueue---------->')
my_queue.enqueue(1)
my_queue.enqueue(2)
my_queue.enqueue(3)
print('\nDequeue------------>')
my_queue.dequeue()
print('\nisFull------------->')
my_queue.isfull()
print('\nisEmpty------------>')
my_queue.is_empty()  
print('\nSize--------------->')
my_queue.size() 
```
